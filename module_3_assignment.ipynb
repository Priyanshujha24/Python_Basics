7. What is the difference between “==” and “is” operators in Python ?

Ans. The "==" operator is a comparison operator which check if the two values are equal or not. Whereas "is" operator is used to check if wto variables pointing to a same memory location. Both returns True or False.

8. What are logical operators in Python ?

Ans. There are 3 main logical operators 'AND', 'OR', 'NOT'. It is commonly used to work with conditional sentences. AND operator returns true only if both the operands are true. OR operator returns true if any of the operand is true. NOT oprator works with only one operand and returns opposite of the operand, if the operand is true it will return false.

9. What is type casting in Python ?

Ans. Type casting in Python refers to the process of explicitly change of data type according to the requirements.This is done using built-in function. For example, 'input' method take data as string from user and if we need the data in 'integer' form we will type cast the data or change its type.

10. What is the difference between implicit and explicit type casting ?

Ans. Implicit type casting means assigning the value of any data type without user intervention. for exapmle : adding one integer to a float value.
Explicit type casting is converting a data type into another required type by user intervention with the help of mentioning the keywords.

12. How does the elif statement work ?

Ans. elif statement works when 'if' statement doesn't meet the expressed condition. It gets executed before else statement.

Ans. 14. Describe a scenario where a while loop is more suitable than a for loop.

```python
#13. Write a Python program to find the sum of all even numbers between 1 and 50

sum = 0
for i in range(2,51):
  if i%2==0:
    sum+=i
print(f"sum of first 50 even number is : {sum}")
```

```python
#14. Write a program to reverse a string using a while loop
name = "priyanshu"
reversed_name = ""
length = len(name)
while length>=1:
  reversed_name += name[length-1]
  length-=1
print(reversed_name)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter number to calculate factorial : "))
fact = 1
while num>0:
  fact=fact*num
  num-=1
print("Factorial of the number is : ",fact)
```
